```python
import json
import tkinter as ttk
from tkinter import *
```

```python
class Chatbot:
    def __init__(self, root, medical_data):
        self.root = root
        self.root.geometry("600x600")
        self.root.title('CHATBOT')
        self.medical_data = medical_data

        # Create a frame for the main content
        main_frame = Frame(self.root, bd=2, relief=RAISED, bg='green')
        main_frame.pack(fill=BOTH, expand=True)

        # Title
        lbl_title = Label(main_frame, text=" Health Related ChatBot", font=('times new roman', 20, 'bold'))
        lbl_title.pack(pady=10)

        # Create a text widget for the chat history
        self.scroll_y = ttk.Scrollbar(main_frame, orient=VERTICAL)
        self.text = Text(main_frame, width=65, height=20, font=('arial', 14), relief=RAISED, yscrollcommand=self.scroll_y.set)
        self.scroll_y.pack(side=RIGHT, fill=Y)
        self.text.pack()

        # Entry field for user input
        self.entry = ttk.Entry(main_frame, font=('times new roman', 15, 'bold'))
        self.entry.pack(fill=X, padx=10, pady=10)

        # Send button
        self.btn_send = ttk.Button(main_frame, text='Send', font=('times new roman', 14, 'bold'), bg='black', fg='white', command=self.send)
        self.btn_send.pack(pady=5)
        
         # Clear button
        self.btn_clr = ttk.Button(main_frame, text='Clear', font=('times new roman', 14, 'bold'), bg='black', fg='white', command=self.clear_text)
        self.btn_clr.pack(pady=5)

    def send(self):
        user_input = self.entry.get()
        self.text.insert(END, f"\n\nYou: {user_input}")

        # Check if user input matches any diseases in the medical data
        for disease in self.medical_data["diseases"]:
            if user_input.lower() == disease["name"].lower():
                self.text.insert(END, f"\n\nChatBot: {disease['description']}")
                self.text.insert(END, "\n\nSymptoms: " + ', '.join(disease['symptoms']))
                self.text.insert(END, "\n\nTreatments: " + ', '.join(disease['treatments']))
                break
        else:
            self.text.insert(END, "\n\nChatBot: I'm not familiar with that disease. Please ask about a known disease.")

    def clear_text(self):
        self.text.delete(1.0, END) 

if __name__ == "__main__":
    root = Tk()
    with open('health-data.json', 'r') as file:
        medical_data = json.load(file)

    chatbot_app = Chatbot(root, medical_data)
    root.mainloop()
```
